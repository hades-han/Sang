```python
# Import thư viện
from PIL import Image
```

```python
# Thư mục chứa ảnh 
# trong python dùng dấu / \\
# Thư mục chứa ảnh
directory = 'C:/HinhAnhThiGiacMayTinh'
```

```python
image_path =  directory + '/img01.png'
print(image_path)
```

```python
img01 = Image.open(image_path)
```

```python
display(img01)
```

```python
print("Định dang ảnh: ", img01.format)
```

```python
print("Kích thước hình ảnh: ", img01.size)
```

```python
# Convert the image to RGB mode if it has an alpha channel
if img01.mode == 'RGBA':
    img01 = img01.convert('RGB')

# Save the image in JPEG format
new_img_01_path = directory + '/new_01.JPG'
new_img_png_01_path = directory + '/new_01.PNG'
img01.save(new_img_01_path)
img01.save(new_img_png_01_path)
```

```python
# Đóng tập tin
img01.close()
```
