```python
# Thư viện
from PIL import Image
from imgtools import *
```

```python
my_path = 'C:/HinhAnhThiGiacMayTinh/img03.png'
```

```python
# Đọc ảnh
img = load_image(my_path)
```

```python
display(img)
```

```python
if img.mode != 'RGB':
    img = img.convert('RGB')
```

```python
# Chuyển ảnh RGB sang ảnh có màu xám
gray_image = img.convert("L")
display(gray_image)
```

```python
# Phân RGB thành 3 tầng màu khác nhau
red_band, green_band, blue_band = img.split()
display(red_band)
display(green_band)
display(blue_band)
```

```python
# Trộn ảnh
merged_image =  Image.merge("RGB", (red_band, green_band, blue_band))
display(merged_image)
```
