```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Đường dẫn ảnh
my_path = 'C:/HinhAnhThiGiacMayTinh/img03.png'
# Đọc ảnh
im = Image.open(my_path)
plt.imshow(im)
```

```python
# Chuyển sang ảnh xám
im_gray = im.convert('L')
plt.imshow(im_gray, cmap='gray')
```

```python
im_array = np.array(im_gray)
print(im_array)
```

```python
# Tạo một hình vẽ mới
plt.figure()
plt.hist(im_array.flatten(),bins=128)
plt.title('Biểu đồ Histogram của ảnh')
plt.xlabel('Giá trị của pixel')
plt.xlabel('Số lượng')
plt.show()
```

```python
# Làm nổi bật đường biên
plt.figure()
plt.gray()
plt.contour(im_gray, origin="image")
plt.show()
```

```python
#Đường dãn ảnh
my_path = 'C:/HinhAnhThiGiacMayTinh/img03.png'
# Đọc ảnh và hiển thị trục tọa độ
im = Image.open(my_path).convert("L")
plt.imshow(im)
# Tạo một cái hình vẽ mới
plt.figure()
# Đặt màu sắc
plt.gray()
# Vẽ biểu đồ đường biên
plt.contour(im, origin="image")
# Đảm bảo tỷ lệ chia trục x và trục y trên biểu đồ là bằng nhau
plt.axis("equal")
# Hiển thị ảnh
plt.show()
```
